```python
## Load the Wage CSV dataset

import pandas as pd

wage_data = pd.read_csv("Wage.csv", index_col=0)
wage_data.head(10)
```

```python
## Display the number of features and thir names 

wage_data.shape[1], list(wage_data.columns)
```

```python
## Delete "lowage" and dsiplay number of eatures and their names again 

wage_data = wage_data.drop(columns=["logwage"])
wage_data.shape[1], list(wage_data.columns)
```

```python
## Display the number of data points 

wage_data.shape[0]
```

```python
## Dsiplay the data in a table "subset"

wage_data.head(15)
```

```python
## Statistic summary of the features 

import numpy as np
import pandas as pd

for column_name in wage_data.columns:
    print(column_name)
    column_series = wage_data[column_name]

    if pd.api.types.is_numeric_dtype(column_series):
        numeric_values = column_series.dropna().to_numpy()
        q1, med, q3 = np.percentile(numeric_values, [25, 50, 75])

        print(f"   Min.   : {numeric_values.min():.6g}")
        print(f"   1st Qu.: {q1:.6g}")
        print(f"   Median : {med:.6g}")
        print(f"   Mean   : {numeric_values.mean():.6g}")
        print(f"   3rd Qu.: {q3:.6g}")
        print(f"   Max.   : {numeric_values.max():.6g}")
    else:
        counts = column_series.value_counts(dropna=False)
        for level, count in counts.items():
            print(f"   {level}: {count}")

    print()
```

```python
## Numerical features 

import matplotlib.pyplot as plt 
from scipy import stats
```

```python
## Age numerical feature 
## Plot the feature against the lable values
x_age = wage_data["age"]
y_wage = wage_data["wage"]

plt.figure()
plt.scatter(x_age, y_wage, alpha=0.2) 

slope, intercept, _, _, _, = stats.linregress(x_age, y_wage)
x_line = np.linspace(x_age.min(), x_age.max(), 300)
y_line = intercept + slope * x_line

plt.plot(x_line, y_line, color="C2", linewidth=10, alpha=0.1)
plt.plot(x_line, y_line, color="C2", linewidth=2, alpha=1.0) 

plt.xlabel("age")
plt.ylabel("wage")
plt.title("age vs wage")
plt.show()
```

```python
## Age numerical feature 
## Test the normality of the feature and label values

shapiro_age_W, shapiro_age_p = stats.shapiro(x_age)
shapiro_wage_W, shapiro_wage_p = stats.shapiro(y_wage)

print(f"Shapiro age: W={shapiro_age_W:.5f}, p={shapiro_age_p:.3g}")
print(f"Shapiro wage: W={shapiro_wage_W:.5f}, p={shapiro_wage_p:.3g}")

plt.figure()
stats.probplot(x_age, dist="norm", plot=plt)
plt.gca().lines[0].set_alpha(0.2)
plt.ylim(None, 100)
plt.ylabel("Ordered values: age")
plt.title("Q-Q plot: age")
plt.show()

plt.figure()
stats.probplot(y_wage.dropna(), dist="norm", plot=plt)
plt.gca().lines[0].set_alpha(0.2)
plt.ylabel("Ordered values: wage")
plt.title("Q-Q plot: wage")
plt.show()
```

```python
## The Pearson correlation

r_age, p_age = stats.pearsonr(wage_data["age"], wage_data["wage"])
print(f"Pearson (age vs wage): r={r_age:.7f}, p={p_age:.7g}")
```

Interpretation (age): Weak positive relationship, Pearson correlation is significant (small p-value), but the effect is not strong "weak 0.20".


```python
## Year numerical feature

x_year = wage_data["year"]
y_wage = wage_data["wage"]

plt.figure()
plt.scatter(x_year, y_wage, alpha=0.2)

slope, intercept, _, _, _ = stats.linregress(x_year, y_wage)
x_line = np.linspace(x_year.min(), x_year.max(), 300)
y_line = intercept + slope * x_line

plt.plot(x_line, y_line, color="C2", linewidth=10, alpha=0.1)
plt.plot(x_line, y_line, color="C2", linewidth=2, alpha=1.0)

plt.xlabel("year")
plt.ylabel("wage")
plt.title("year vs wage")
plt.show()
```

```python
## Test the normality of the feature and label values

W_year, p_year_norm = stats.shapiro(x_year)
W_wage, p_wage_norm = stats.shapiro(y_wage)

print(f"Shapiro-Wilk year: W={W_year:.6f}, p-value={p_year_norm:.6g}")
print(f"Shapiro-Wilk wage: W={W_wage:.6f}, p-value={p_wage_norm:.6g}")

plt.figure()
stats.probplot(x_year, dist="norm", plot=plt)
plt.title("Q-Q plot: year")
plt.show()

plt.figure()
stats.probplot(y_wage, dist="norm", plot=plt)
plt.title("Q-Q plot: wage")
plt.show()
```

```python
## The Pearson correlation

r_year, p_year = stats.pearsonr(wage_data["year"], wage_data["wage"])
print(f"Pearson (year vs wage): r={r_year:.7f}, p={p_year:.7g}")
```

Interpretation (year): Very weak relationship. Pearson correlation can be significant due to many samples, but the effect is very weak.


```python
## the non-numerical features

np.random.seed(0)
```

```python
## maritl non numerical 

feature_col = "maritl"
levels = wage_data[feature_col].dropna().unique().tolist()
print(levels)
```

```python
## Plotting "box plot"

plt.figure(figsize=(10, 4))
wage_data.boxplot(column="wage", by=feature_col, grid=False, rot=30)
plt.title(f"{"wage"} by {feature_col}")
plt.suptitle("")
plt.xlabel(feature_col)
plt.ylabel("wage")
plt.show()
```

```python
#ANOVA

groups = [wage_data.loc[wage_data[feature_col] == lvl, "wage"].dropna() for lvl in levels]
F_value, p_value = stats.f_oneway(*groups)
print(f"ANOVA wage ~ {feature_col}: F={F_value:.3f}, p-value={p_value:.6g}")
```

Interpretation (maritl): Wage differs between marital status groups, ANOVA p value showing the group means are not all equal.
