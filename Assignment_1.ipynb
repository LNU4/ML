```python
## Numerical features 

import numpy as np 
import matplotlib.pyplot as plt 
from scipy import stats
```

```python
## Age numerical feature 
## Plot the feature against the lable values
x_age = wage_data["age"]
y_wage = wage_data["wage"]

plt.figure()
plt.scatter(x_age, y_wage, alpha=0.2) 

slope, intercept, _, _, _, = stats.linregress(x_age, y_wage)
x_line = np.linspace(x_age.min(), x_age.max(), 300)
y_line = intercept + slope * x_line

plt.plot(x_line, y_line, color="C2", linewidth=10, alpha=0.1)
plt.plot(x_line, y_line, color="C2", linewidth=2, alpha=1.0) 

plt.xlabel("age")
plt.ylabel("wage")
plt.title("age vs wage")
plt.show()
```

```python
## Age numerical feature 
## Test the normality of the feature and label values

shapiro_age_W, shapiro_age_p = stats.shapiro(x_age)
shapiro_wage_W, shapiro_wage_p = stats.shapiro(y_wage)

print(f"Shapiro age: W={shapiro_age_W:.5f}, p={shapiro_age_p:.3g}")
print(f"Shapiro wage: W={shapiro_wage_W:.5f}, p={shapiro_wage_p:.3g}")

plt.figure()
stats.probplot(x_age, dist="norm", plot=plt)
plt.gca().lines[0].set_alpha(0.2)
plt.ylim(None, 100)
plt.ylabel("Ordered values: age")
plt.title("Q-Q plot: age")
plt.show()

plt.figure()
stats.probplot(y_wage.dropna(), dist="norm", plot=plt)
plt.gca().lines[0].set_alpha(0.2)
plt.ylabel("Ordered values: wage")
plt.title("Q-Q plot: wage")
plt.show()
```

```python
## The Pearson correlation

r_age, p_age = stats.pearsonr(wage_data["age"], wage_data["wage"])
print(f"Pearson (age vs wage): r={r_age:.7f}, p={p_age:.7g}")
```

```python
## Interpretation of the results "Age$Wage"

## The scatter plot shows a slight positive relationship between age and wage, where wages tend to increase with age, particularly from the early 20s to the 30s.
## However, the data points are widely spread, indicating that age alone is not a strong predictor of wage.
## High wage values appear mainly at middle and higher ages, suggesting a right-skewed wage distribution.

## The ShapiroWilk test rejects the normality assumption for both variables.
## For age, the QQ plot follows the reference line in the center but deviates in the tails. 
## For wage, the very small p-value and strong curvature in the QQ plot indicate that the distribution is substantially non normal, with heavy tails and outliers.
```

```python
## the non-numerical features

levels = wage_data["education"].unique()
print(levels)

```

```python
plt.figure(figsize=(10, 4))
wage_data.boxplot(column="wage", by="education", grid=False, rot=30)
plt.title("wage by education")
plt.suptitle("")
plt.xlabel("education")
plt.ylabel("wage")
plt.show()
```

```python
feature_col = "education"
levels = wage_data[feature_col].dropna().unique()

groups = [wage_data.loc[wage_data["education"] == lvl, "wage"] for lvl in levels]

F_value, p_value = stats.f_oneway(*groups)
print(f"ANOVA wage ~ education: F={F_value:.1f}, p={p_value:.2g}")
```

```python
## Interpretation (Education vs Wage)

## The box plots show that wages increase with higher levels of education, as reflected by rising median values from non high school graduates to advanced degree holders.
## indicating that higher education is associated with higher earnings. 
## Although there is overlap between groups and some high wage outliers in each category, the overall shift in distributions is obvious.

## The one-way ANOVA test is highly significant (p = 2.9e-172).
## Indicating that the mean wages differ significantly across education segments. Therefore, education has a strong effect on wage.
```
