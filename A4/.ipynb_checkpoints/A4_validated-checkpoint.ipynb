## Conceptual

1. Discuss the differences between LDA and QDA in terms of their main assumptions about classes, decision boundaries, number of samples, and overfitting.

- **LDA (Linear Discriminant Analysis)** typically assumes each class follows a multivariate Normal distribution and that **all classes share the same covariance matrix**. This shared-covariance assumption produces a **linear** decision boundary and requires estimating fewer parameters, so performance is usually more stable when the sample size is limited and the risk of overfitting is lower.
- **QDA (Quadratic Discriminant Analysis)** also typically assumes multivariate Normal class-conditional distributions, but it allows **a separate covariance matrix for each class**. This produces a **quadratic (curved)** decision boundary and requires estimating more parameters, so it can capture more complex class separation patterns but usually needs more data and can overfit more easily when the dataset is not large.


2. Regarding KNN:

a) How does the choice of distance metrics affect the performance of k-NN classification?

- The distance metric defines what “nearest” means, so it determines which training observations become the neighbors used for voting. Different metrics can produce different neighbor sets and therefore different predictions. Feature scaling is important because many distance metrics can be dominated by variables with larger numerical ranges.

b) Please also discuss the concept of the curse of dimensionality and its implications for the k-NN algorithm.

- As the number of predictors increases, points tend to become far apart and distances become less informative (many observations look similarly distant). Because k-NN depends on meaningful local neighborhoods, its performance can degrade in high dimensions unless there is substantially more data or dimensionality is reduced (for example, via feature selection or dimensionality reduction).


## Practical

```python
# Environment check and configuration (run this once at the start)

import sys
from pathlib import Path

import pandas as pandas
import numpy as numpy
import matplotlib
import matplotlib.pyplot as matplotlib_pyplot

import statsmodels
import statsmodels.api as statsmodels_api
import sklearn

from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

print("Python:", sys.version.split()[0])
print("pandas:", pandas.__version__)
print("numpy:", numpy.__version__)
print("matplotlib:", matplotlib.__version__)
print("statsmodels:", statsmodels.__version__)
print("scikit-learn:", sklearn.__version__)

# Dataset location (edit this if Smarket.csv is in a different folder)
smarket_csv_file_path = Path("Smarket.csv")

if not smarket_csv_file_path.exists():
    print("\nERROR: Smarket.csv was not found in the current folder:", Path.cwd())
    print("Put Smarket.csv in the same folder as this notebook, or update smarket_csv_file_path.")
    print("Files in current folder:", [p.name for p in Path.cwd().iterdir()][:50])
    raise FileNotFoundError("Smarket.csv not found. Fix smarket_csv_file_path and re-run this cell.")
```

## Interpretation
This cell verifies that the required libraries are available, prints package versions for reproducibility, and checks that `Smarket.csv` exists in the notebook’s working directory. A clear file-path check prevents silent failures later (for example, file-not-found errors).

```python
# Load the data and get an overview of the data

smarket_dataset = pandas.read_csv(smarket_csv_file_path)

# If there is an extra index-like column, remove it
if "Unnamed: 0" in smarket_dataset.columns:
    smarket_dataset = smarket_dataset.drop(columns=["Unnamed: 0"])

print("Dataset shape (rows, columns):", smarket_dataset.shape)
print("\nColumn names:", smarket_dataset.columns.to_list())
print("\nDirection class counts:\n", smarket_dataset["Direction"].value_counts())
print("\nFirst 5 rows:\n", smarket_dataset.head(5))

numeric_columns_dataframe = smarket_dataset.select_dtypes(include=[numpy.number])
print("\nSummary statistics (numeric columns):\n", numeric_columns_dataframe.describe())

print("\nMissing values per column:\n", smarket_dataset.isna().sum())
```

## Interpretation
This provides a basic data audit: the number of rows/columns, variable names, the class balance for `Direction`, a preview of records, summary statistics for numeric variables, and missing-value counts. These checks confirm the dataset is readable and suitable for classification steps.

```python
# Correlation matrix + plot

predictor_correlation_matrix = numeric_columns_dataframe.corr()
print("Correlation matrix (numeric columns):\n", predictor_correlation_matrix)

matplotlib_pyplot.figure()
matplotlib_pyplot.scatter(smarket_dataset["Year"], smarket_dataset["Volume"])
matplotlib_pyplot.xlabel("Year")
matplotlib_pyplot.ylabel("Volume")
matplotlib_pyplot.title("Year vs Volume")
matplotlib_pyplot.show()
```

## Interpretation
The correlation matrix summarizes linear relationships between numeric variables. A notable relationship is the positive correlation between `Year` and `Volume` (approximately 0.54), and the scatter plot provides a visual confirmation that trading volume generally increases over time.

```python
# Perform a logistic regression
import statsmodels.api as statsmodels_api

predictor_columns_lag1_to_lag5_and_volume = ["Lag1", "Lag2", "Lag3", "Lag4", "Lag5", "Volume"]
predictor_matrix_lag1_to_lag5_and_volume = smarket_dataset[predictor_columns_lag1_to_lag5_and_volume]

direction_binary_response = (smarket_dataset["Direction"] == "Up").astype(int)

predictor_matrix_with_intercept_all_predictors = statsmodels_api.add_constant(
    predictor_matrix_lag1_to_lag5_and_volume,
    has_constant="add"
)

logistic_regression_model_all_predictors = statsmodels_api.Logit(
    direction_binary_response,
    predictor_matrix_with_intercept_all_predictors
).fit(disp=False)

print(logistic_regression_model_all_predictors.summary())
```

## Interpretation
This model estimates how `Lag1`–`Lag5` and `Volume` relate to the probability that `Direction = Up`. The coefficient table indicates which predictors have stronger evidence of association with the response; weak significance across predictors typically suggests limited predictive signal for this task using these variables alone.

```python
# Use the logistic regression models
from sklearn.metrics import confusion_matrix, accuracy_score

predicted_probabilities_up_all_predictors = logistic_regression_model_all_predictors.predict(
    predictor_matrix_with_intercept_all_predictors
)

predicted_class_labels_all_predictors = (predicted_probabilities_up_all_predictors >= 0.5).astype(int)

confusion_matrix_all_predictors = confusion_matrix(
    direction_binary_response,
    predicted_class_labels_all_predictors
)

accuracy_all_predictors = accuracy_score(direction_binary_response, predicted_class_labels_all_predictors)

true_negatives = confusion_matrix_all_predictors[0, 0]
false_positives = confusion_matrix_all_predictors[0, 1]
false_negatives = confusion_matrix_all_predictors[1, 0]
true_positives = confusion_matrix_all_predictors[1, 1]

precision_value = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else numpy.nan
recall_value = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else numpy.nan
specificity_value = true_negatives / (true_negatives + false_positives) if (true_negatives + false_positives) > 0 else numpy.nan

print("Confusion matrix (rows=true, cols=pred):\n", confusion_matrix_all_predictors)
print("Accuracy:", accuracy_all_predictors)
print("Precision:", precision_value)
print("Recall (Sensitivity):", recall_value)
print("Specificity:", specificity_value)
```

## Interpretation
The confusion matrix shows correct and incorrect classifications for `Down` (0) and `Up` (1). For this model, accuracy is 0.5216, which is only slightly above a majority-class baseline. This indicates that, with a 0.5 threshold, the fitted logistic model provides limited improvement over a simple default prediction rule.

```python
# Baseline, null classifier
most_frequent_class_label = int(direction_binary_response.mode().iloc[0])
null_classifier_predictions = numpy.full(shape=len(direction_binary_response), fill_value=most_frequent_class_label)

null_classifier_confusion_matrix = confusion_matrix(direction_binary_response, null_classifier_predictions)
null_classifier_accuracy = accuracy_score(direction_binary_response, null_classifier_predictions)

print("Null classifier confusion matrix:\n", null_classifier_confusion_matrix)
print("Null classifier accuracy:", null_classifier_accuracy)
```

## Interpretation
A null classifier that always predicts the most frequent class (`Up`) achieves accuracy 0.5184. This provides a reference point: a useful model should meaningfully exceed this baseline rather than only improving by a fraction of a percent.

```python
# Logistic regression with subset Lag1, Lag2 "fit and evaluate"

predictor_columns_lag1_and_lag2 = ["Lag1", "Lag2"]
predictor_matrix_lag1_and_lag2 = smarket_dataset[predictor_columns_lag1_and_lag2]

predictor_matrix_with_intercept_lag1_and_lag2 = statsmodels_api.add_constant(
    predictor_matrix_lag1_and_lag2,
    has_constant="add"
)

logistic_regression_model_lag1_and_lag2 = statsmodels_api.Logit(
    direction_binary_response,
    predictor_matrix_with_intercept_lag1_and_lag2
).fit(disp=False)

predicted_probabilities_up_lag1_and_lag2 = logistic_regression_model_lag1_and_lag2.predict(
    predictor_matrix_with_intercept_lag1_and_lag2
)

predicted_class_labels_lag1_and_lag2 = (predicted_probabilities_up_lag1_and_lag2 >= 0.5).astype(int)

confusion_matrix_lag1_and_lag2 = confusion_matrix(direction_binary_response, predicted_class_labels_lag1_and_lag2)
accuracy_lag1_and_lag2 = accuracy_score(direction_binary_response, predicted_class_labels_lag1_and_lag2)

print(logistic_regression_model_lag1_and_lag2.summary())
print("\nConfusion matrix (Lag1+Lag2):\n", confusion_matrix_lag1_and_lag2)
print("Accuracy (Lag1+Lag2):", accuracy_lag1_and_lag2)
```

## Interpretation
Using only `Lag1` and `Lag2` simplifies the model and reduces the number of estimated parameters. The resulting accuracy is 0.5280, which is slightly higher than the all-predictor logistic model, suggesting the additional predictors may not add much predictive value in this dataset.

```python
# LDA with Lag1 and Lag2 "fit and evaluate"

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

linear_discriminant_analysis_model_lag1_and_lag2 = LinearDiscriminantAnalysis()
linear_discriminant_analysis_model_lag1_and_lag2.fit(
    predictor_matrix_lag1_and_lag2,
    direction_binary_response
)

predicted_class_labels_lda = linear_discriminant_analysis_model_lag1_and_lag2.predict(
    predictor_matrix_lag1_and_lag2
)

confusion_matrix_lda = confusion_matrix(direction_binary_response, predicted_class_labels_lda)
accuracy_lda = accuracy_score(direction_binary_response, predicted_class_labels_lda)

print("LDA confusion matrix:\n", confusion_matrix_lda)
print("LDA accuracy:", accuracy_lda)
```

## Interpretation
LDA forms a linear decision rule based on class means and a shared covariance estimate. The accuracy is 0.5280, matching the `Lag1`+`Lag2` logistic regression result here, indicating similar linear-separation performance for these two predictors.

```python
# QDA with Lag1 and Lag2 "fit and evaluate"
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

quadratic_discriminant_analysis_model_lag1_and_lag2 = QuadraticDiscriminantAnalysis()
quadratic_discriminant_analysis_model_lag1_and_lag2.fit(
    predictor_matrix_lag1_and_lag2,
    direction_binary_response
)

predicted_class_labels_qda = quadratic_discriminant_analysis_model_lag1_and_lag2.predict(
    predictor_matrix_lag1_and_lag2
)

confusion_matrix_qda = confusion_matrix(direction_binary_response, predicted_class_labels_qda)
accuracy_qda = accuracy_score(direction_binary_response, predicted_class_labels_qda)

print("QDA confusion matrix:\n", confusion_matrix_qda)
print("QDA accuracy:", accuracy_qda)
```

## Interpretation
QDA allows a separate covariance estimate for each class, producing a curved decision boundary in the predictor space. The accuracy is 0.5304, slightly higher than the linear methods in this case, which suggests that a modest amount of non-linearity may be present (while still indicating overall weak predictability).

```python
# k-NN, train on Year < 2005, test on Year == 2005
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

training_rows_mask_year_less_than_2005 = smarket_dataset["Year"] < 2005

training_predictors_lag1_and_lag2 = smarket_dataset.loc[training_rows_mask_year_less_than_2005, ["Lag1", "Lag2"]]
training_direction_binary_response = direction_binary_response.loc[training_rows_mask_year_less_than_2005]

testing_predictors_lag1_and_lag2 = smarket_dataset.loc[~training_rows_mask_year_less_than_2005, ["Lag1", "Lag2"]]
testing_direction_binary_response = direction_binary_response.loc[~training_rows_mask_year_less_than_2005]

standard_scaler_for_knn = StandardScaler()
training_predictors_scaled = standard_scaler_for_knn.fit_transform(training_predictors_lag1_and_lag2)
testing_predictors_scaled = standard_scaler_for_knn.transform(testing_predictors_lag1_and_lag2)

for number_of_neighbors in [1, 3]:
    knn_classifier_model = KNeighborsClassifier(n_neighbors=number_of_neighbors)
    knn_classifier_model.fit(training_predictors_scaled, training_direction_binary_response)

    predicted_testing_class_labels = knn_classifier_model.predict(testing_predictors_scaled)

    confusion_matrix_knn = confusion_matrix(testing_direction_binary_response, predicted_testing_class_labels)
    accuracy_knn = accuracy_score(testing_direction_binary_response, predicted_testing_class_labels)

    print("\nk-NN with number_of_neighbors =", number_of_neighbors)
    print("Confusion matrix:\n", confusion_matrix_knn)
    print("Accuracy:", accuracy_knn)
```

## Interpretation
This evaluates generalization using a time split: training years before 2005 and testing on 2005. For `k=1`, accuracy is 0.5000, which is essentially chance-level. For `k=3`, accuracy is 0.5317, showing a small improvement when voting uses a slightly larger neighborhood. Minor differences from other software implementations can occur due to tie-breaking rules.

```python
# ROC curve, AUC, and a custom threshold (uses logistic regression Lag1+Lag2 probabilities)

false_positive_rate_values, true_positive_rate_values, threshold_values = roc_curve(
    direction_binary_response,
    predicted_probabilities_up_lag1_and_lag2
)

area_under_curve_value = roc_auc_score(
    direction_binary_response,
    predicted_probabilities_up_lag1_and_lag2
)

matplotlib_pyplot.figure()
matplotlib_pyplot.plot(false_positive_rate_values, true_positive_rate_values)
matplotlib_pyplot.xlabel("False Positive Rate")
matplotlib_pyplot.ylabel("True Positive Rate")
matplotlib_pyplot.title("ROC Curve (Logistic Regression: Lag1 + Lag2)")
matplotlib_pyplot.show()

print("AUC:", area_under_curve_value)

custom_threshold_value = 0.20
predicted_class_labels_threshold_020 = (predicted_probabilities_up_lag1_and_lag2 >= custom_threshold_value).astype(int)

print("Confusion matrix at threshold 0.20:\n",
      confusion_matrix(direction_binary_response, predicted_class_labels_threshold_020))
```

## Interpretation
The ROC curve summarizes how the true-positive rate and false-positive rate trade off across thresholds. For the `Lag1`+`Lag2` logistic model, AUC is approximately 0.531; values near 0.5 indicate weak separation between classes, so this AUC suggests only limited ranking ability. Lowering the classification threshold increases predicted `Up` frequency but typically raises false positives.
